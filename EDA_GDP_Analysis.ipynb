### Calculate the average growth of states over the duration 2013-14, 2014-15 and 2015-16 by taking the mean of the row '(% Growth over previous year)'. Compare the calculated value and plot it for the states.

```python

#Getting only required data from Duration column
df_state = df_csv.loc[((df_csv.Duration == '2013-14') | (df_csv.Duration == '2014-15') | (df_csv.Duration == '2015-16')) & (df_csv['Items  Description'] == '(% Growth over previous year)')]

# Data Cleaning as no need of Item Description,All_india GDP Column and West Bengal as it contains all NAN value
df_state = df_state.drop(['Items  Description','West Bengal1'], 1)

#Find the Mean across all the rows
df_state= df_state.append(df_state.iloc[:, :].mean(axis=0), ignore_index=True)

#Replace the calculated value with name as 'Mean'
df_state[['Duration']] = df_state[['Duration']].fillna(value='Mean')

#Setting the Duration as index
df_state.set_index('Duration',inplace=True)

# To better visualization sort value in decresing order
df_state = df_state.sort_values(by ='Mean',axis=1) 
#df

#No need of All_india GDP column and getting to other dataframe for plotting
df_growth = df_state.drop('All_India GDP', 1)

#Plotting the figure using seaborn library x axis for states and y axis for mean
plt.figure(figsize=(15,8))
sns.barplot(y = df_growth.loc['Mean'], x = df_growth.columns)
plt.ylabel('Average Growth Rate', fontsize=12)
plt.xlabel('States', fontsize=12)
plt.xticks(rotation='vertical')
plt.show()


```

```python
def heatmap(x, y, size):
    fig, ax = plt.subplots()
    
    # Mapping from column names to integer coordinates
    x_labels = [v for v in sorted(x.unique())]
    y_labels = [v for v in sorted(y.unique())]
    x_to_num = {p[1]:p[0] for p in enumerate(x_labels)} 
    y_to_num = {p[1]:p[0] for p in enumerate(y_labels)} 
    
    size_scale = 500
    ax.scatter(
        x=x.map(x_to_num), # Use mapping for x
        y=y.map(y_to_num), # Use mapping for y
        s=size * size_scale, # Vector of square sizes, proportional to size parameter
        marker='s' # Use square as scatterplot marker
    )
    
    # Show column labels on the axes
    ax.set_xticks([x_to_num[v] for v in x_labels])
    ax.set_xticklabels(x_labels, rotation=45, horizontalalignment='right')
    ax.set_yticks([y_to_num[v] for v in y_labels])
    ax.set_yticklabels(y_labels)
    

corr_2014_15 = pd.melt(corr.reset_index(), id_vars='index') # Unpivot the dataframe, so we can get pair of arrays for x and y
corr_2014_15.columns = ['x', 'y', 'value']
heatmap(
    x=corr_2014_15['x'],
    y=corr_2014_15['y'],
    size=corr_2014_15['value'].abs()
)


```
